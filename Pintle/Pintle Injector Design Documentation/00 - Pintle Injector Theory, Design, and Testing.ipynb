The Portland State Aerospace Society is in the process of testing its $2.2kN$ ($500 lb_f$) regeneratively cooled bi-propellant rocket motor. One of the critical steps prior to hot firing the motor is the testing and validating of the propellant injector. For this motor, a pintle style injector was selected and an aluminum test article was machined according to the designs provided by the 2015 Capstone team.  

## PSAS's first bi-prop engine attempt


The 2015 engine was designed to employ a liquid oxygen (LOX) centered pintle design. In this design, LOX flows through the center of the pintle and is sprayed outward in a radial pattern. The fuel enters a chamber surrounding the base of the pintle and enters the engine through the annular gap between the outer radius of the pintle and the engine. (See Figures 1 & 2) Inside of the combustion chamber, the two propellants impinge upon each other at right angles. This, ideally, results in a cone shaped spray of well mixed fuel and oxidizer.

# Theory


## Operational Theory

The pintle injector serves two primary purposes in the engine. First, it delivers the fuel and the oxidizer to the combustion chamber at the correct mass flow rate and pressure. Secondly it mixes the propellants and promotes proper combustion.

In practice, the mass flow rate through the injector is controlled by the input pressure for the fuel and oxidizer upstream of the injector. Ideally, the injector will have a pressure drop of between 15 and 25 % of the engine's chamber pressure. 

>This drop isolates chamber-pressure oscilations from the feed system, reducing pressure coupling between the combustion chamber and the feed system which could lead to instabilities or oscillations in the flow that are driven by various in combustion. [1]

For pintle injectors, this pressure drop can potentially go as low as 5% [1]. Until PSAS has built some heritage with our pintle designs, it is recommended that the 15-25% range be aimed for. 

According to some of the available research on pintle injectors, the optimum propellant spray angle is $45^\circ$. [2]

## Key design Equations 

### Pressure loss coefficient 

The pressure loss coefficient is a key parameter in the design of injectors. With the pressure loss coefficient known, it is possible to determine the mass flow rate of each propellant through the injector by measuring the pressure directly upstream of the injector and in the chamber.




The governing equations employed for the first approximations in this design were:


Mass flow rate:

\begin{equation}
 \dot{m} = \rho VA
\label{eq:massflow}
\tag{1}
\end{equation} 

Where;

$\dot{m} =$  Mass Flow Rate (from flow meter) 

$\rho = $ Density of working fluid 

$V = $ Mean fluid velocity

$A = $ Flow area

Given that high Reynolds numbers are expected, it is being assumed that the pressure losses will be dominated by the inertial effects, rather than viscous effects.

With this assumption, the energy equation can have the elevation, velocity, and head loss due to friction neglected, which leaves:

\begin{equation}
P_1-P_2 = K_{L}\frac{\rho V_2^2}{2}
\label{eq:energy}
\tag{2}
\end{equation}

Where;

$K_L = $ Pressure Loss coefficient  
$P_1-P_2 =$ Pressure drop 


Equation 1 and 2 can be combined by substituting $V = \frac{\dot{m}}{\rho  A}$ and rearranging to give;

\begin{equation}
\dot{m} = AC_d \sqrt{2\rho(\Delta P)}
\label{eq:massflow cd}
\tag{3}
\end{equation}

which can be rewritten as;

\begin{equation}
C_d = \frac{\dot{m}}{A \sqrt{2\rho(\Delta P)}}
\label{eq:Cd}
\tag{4}
\end{equation}

Where;

$C_D = $ Coefficient of discharge for Pintle and Annulus

$A =$ Flow area(s) of Pintle or Annulus

$\dot{m} =$  Mass Flow Rate (from flow meter) 

$\Delta P =$ Pressure upstream of Pintle or Annulus (gauge) - atmospheric pressure (0 psi)

$\rho = $ Density of water (calculate based on temperature)


For the purposes of this loss coefficient, the area being used will be the exit area of the injector.

Many references, but not all, use $C_D$ for the loss coefficient for injectors as shown in equation \ref{eq:Cd}. As a matter of preference, the loss coefficient equations used within the jupyter notebook codes developed on this project have used the following form of the loss coefficient equations: 


\begin{equation}
K_L = \frac{2\Delta P}{\rho \big(\frac{\dot{m}}{\rho A}\big)^2}
\label{eq:Kl}
\tag{7}
\end{equation}

or


\begin{equation}
K_L = \frac{2\Delta P}{\rho V^2}
\label{eq:Kl alt}
\tag{8}
\end{equation}

where;  $ K_L = \frac{1}{\sqrt{C_d}}$ 

Equation \ref{eq:Kl} shows that in order to empirically determine $K_L$, one must know  the exit area(s) of the injector, the mass flow rate and the corresponding pressure drop.

The exit orifices can be precisely measured using gauge pins, and the mass flow rate and pressure drop can be measured in controlled experiments. 



### Momentum Ratio

According to available research, the spray angle of the mixed propellants has an effect on engine performance.[2] Optimum combustion is supposed to occur when the spray angle is approximately $45^\circ$. Theoretically this happens when the ratio of the momentum rates of the two propellants is equal to 1. 

This ratio can be expressed as;


\begin{equation}
\frac{\dot{m_f}V_f}{\dot{m_o}V_o} = 1
\label{eq:momentum}
\tag{9}
\end{equation}

By substituting the velocity in the momentum ratio for the mass flow rate as defined in equation \ref{eq:massflow}, the optimum momentum ratio equation can be rewritten as;


\begin{equation}
\frac{\dot{m_f}^2}{\rho_f A_f} = \frac{\dot{m_o}^2}{\rho_o A_o}
\label{eq:momentum alt}
\tag{10}
\end{equation}


The calculations for the momentum ratio are slightly non-intuitive because the area used in the calculation is not always the entire exit area.

For the V1, V2, and V3 pintle designs, the exit areas used to calculate for the momentum ratios were:

Pintle: $\pi \frac{d_p^2}{4}$

Annulus: $d_p\times d_g$

Where; $d_p$ is equal to the diameter of the exit orifices of the pintle, and $d_g$ is equal to the annular gap defined as:

$d_g = \frac{1}{2}(A_{OD} - P_{OD})  $

Where $A_{OD}$ is the outer diameter of the Annulus, and $P_{OD}$ is the outer diameter of the pintle.

- Should probably include a drawing or diagram showing the area's involved


### Weber Number

The Weber number is the ratio between inertial forces and surface tension.

\begin{equation}
\frac{\rho V^2 D}{\sigma}
\label{eq:Weber}
\tag{12}
\end{equation}

Where; $L$ is a characteristic length (exit diameter of holes, or annular gap) $\sigma$ is the surface tension

The weber number is an important parameter in the formation of droplets. When a fluid jet has a high weber number at its exit, the inertial forces cause the fluid to break up into small droplets. This is the basic principal behind spray bottle atomizers. 

It is suggested that the exit velocity of LOX remain above 10 m/s in order to obtain a sufficiently high Weber number for acceptable droplet formation. [3] The fuel should seek to have a weber number at least as high as that of the oxidizer.


### Reynolds Number
- justifies use of turbulent modeling for CFD

\begin{equation}
\frac{\rho V D}{\mu}
\label{eq:Reynolds}
\tag{11}
\end{equation}

Where; $D$ is a characteristic length (exit diameter of holes, or the hydraulic diameter of the annulus) $\sigma$ is the surface tension.



# Design 

## Reality isn't always convenient

### Loss coefficients

Determining the loss coefficients for the a pintle style injector was one of the first challenges to be overcome in designing an injector which met both the mass flow rate and pressure requirements of PSAS's first rocket engine.

The initial strategy was to break the internal flow paths of the injector into sections and estimate loss coefficients for each section. 

<img src="./Images/LOX_flow.PNG" align="center" width="500"/>

> Figure 4: Cross section view of V1 pintle with locations of initial loss coefficients used for LOX flow path

<img src="./Images/Fuel_flow.PNG" align="center" width="500"/>

> Figure 5: Cross section view of V1 pintle with locations of initial loss coefficients used for fuel flow path.

The idea was that each flow path within the injector has multiple changes to flow area between point of measurement and the atmosphere. The early hand calculations in this project attempted to approximated the overall loss coefficient by segregating the internal flow geometry into segments which could examined individually for pressure losses. 

The sum of those pressure losses could then be added together to determine a total pressure drop estimate, $\Delta P$. This $\Delta P$ could then be used along with the mass flow rate to estimate a total $K_L$ for the injector. This showed early promise with the V2 injector and gave pressure drops which agreed with CFD models and physical testing. When attempts were made to scale up the injector design to meet the general requirements of an engine capable of reaching 100km, the errors in the estimate quickly diverged from CFD models. 




- A note of caution, hand calculations and CFD should be employed during the design phase, but any injector which is to be used in an engine needs be specifically tested to determine its loss coefficients and spray angle performance. Small variations in hole sizes, even when within tolerance, can effect the actual performance of an injector. Theoretical models are valuable to steer the design into the right neighborhood, but still require physical testing for validation.

- Loss coefficients for this design don't scale linearly

## Tested Designs

### Known issues and how they were addressed

#### V1

This design was undersized. The combined flow rate of water for this version was approximately 3 gpm, while the design calls for nearly 15 gpm. 

#### V2 

This design met the flow rate requirements. It was not tested for spray angle because there was visibly uneven flow through the annulus. The cause of this was suspected to be flow separation in the fuel chamber which did not have adequate time to fully develop inside of the annulus. The solution to this was to split the flow into the fuel chamber into three, radially spaced flow paths.  


#### V3

- Hydro testing data from the V3 shows that the flow rates for the injector are within the specified ranges
- The changes to the fuel flow path worked as intended and there was no noticeable annular flow issues.

- Issue: The mass flow rates needed to create a $45^\circ$ spray angle would not give a momentum ratio of 1. We would have to be even more fuel rich in order to obtain the correct spray angle, but then we would not have the correct mixture ratio.

#### V4 Design for 2.2 kN

### The LV4 injector as a scaled up 2.2kN design

The first attempt to explore an injector design within the approximate flow range of which may be needed by LV4 yielded interesting results. By interesting I mean frustrating and initially confusing.

This first attempt sought to scale up the V3 2.2kN injector on the naive assumption that the loss coefficients and exit velocities would scale in a predictable enough fashion to be able to roughly calculate the correct size based upon the physical testing and CFD results for the 2.2kN injectors. This assumption was incorrect.

As stated earlier, both the loss coefficients and the exit velocities did not behave as originally assumed.  This prompted a series of full factorial experiments using CFD, where key dimensions and parameters were systematically adjusted to isolate the key parameters which affect the loss coefficients for this design. The results of the experiments were then used to create a set of piecewise curve fit equations that can be used to obtain a good estimate of the loss coefficients for any size injector between the V1 2.2kN injector with approximately 0.2 kg/s total propellant mass flow rate, up to an injector with over 2.5 kg/s total propellant mass flow rate.

It was while doing this analysis and experimenting that it became apparent that the original pintle design will not work for a 6 or 8 kN engine without some modifications to address the issues created by flow separation in the radial orifices.

# Manufacturing

## Test articles

Any new injector needs to be tested and validated prior use in a engine. Early physical testing can and should be done with water in order to experimentally verify the loss coefficients and spray angle. 

This is especially true when exploring a new injector design, or a design which has been altered to obtain a different mass flow rate or mixture ratio. 

This creates a manufacturing challenge due to the material and required tolerances which are needed to manufacture an injector. During the design phase, attention should be given to finding ways to produce dimensionally accurate, lower cost, prototypes which can be made, tested, and iterated upon **quickly**.

The fast, low the cost of the prototypes allow for faster development of a functioning injector design which will meet an engines requirements.

As of the 2018-2019 capstone several different prototyping methods have been attempted in the development of the 2.2kN V series injectors.


Pintle injectors have certain dimensions which require very high precision in order to function properly. 

Ensuring that the pintle is correctly centered in the annulus is particularly important. Even small amounts of concentricity error can lead to uneven exit velocity, a lopsided spray angle, and less than optimum combustion.

After any new injector design has been manufactured, careful measurements should be taken of the exit orifices and gaps prior to hydro testing. As-built dimensions should be recorded for later use in determining the loss coefficients.  


### Machined Aluminum


The V1 and V2 injector assemblies were both manufactured in aluminum. These required students who had machining experience in order to make and were time consuming to produce. They required very careful machining and assembly. 

The particular design style for these injectors has room for improvement when it comes to manufacturability. 

Aluminum is a relatively soft metal, which made it very easy to accidentally strip any internal threads which were tapped into the aluminum. This particular issue could be solve by using helicoil threated inserts for any tapped holes.  

Also, the use of shoulder bolts to center the pintle was tried for both the V1 and V2 pintles. The V3 pintle employed countersunk screw heads which made centering the pintle into the annulus much easier to accomplish during assembly.

## Final injector

The first major concern for the final version of an injector is material selection. The injector is going to be subjected to both liquid oxygen, high temperature combustion products, and several hundred PSI. The tip of the pintle is exposed to extreme heating which can potentially damage the pintle and make it a single use item. Thought should be put into how to either make the pintle tip easily replaced between tests, or how to keep it from getting too hot, preferably both.

The metal for the pintle needs to be LOX compatible: Generally this means 304 or 316 Stainless steel, with 304 having a slightly higher melting temperature. There are other material options as well such as Inconel (1390-1424$^\circ$C), Nickel, Copper, or Nickel-Copper Alloys.  

Aluminum is LOX compatible, but will not take the extreme heat which is expected on the pintle tip. (melting point ~ 660$^\circ$C)

For example it may be possible to 3D print the injector in Inconel. It would have its own manufacturing challenges when it comes to post processing, but it could potentially be done.


# Hydro Testing

## Purpose and Theory 

Once an injector design has been manufactured and checked to ensure all of the components fit properly, the first round of testing is performed with water.


## Deliverables

### Loss Coefficients
The first key deliverable from this test is to determine the loss coefficient for each propellant flow path through the injector.

### Spray Angle 
The second deliverable is verification that the injector will create a 45$^\circ$ spray pattern with acceptable mixing and droplet formation.

### Momentum Ratio
The third deliverable is determining if the injector meets the momentum ratio requirement for the propellant mass flow rates.  


## Testing equipment

- Pressure transmitters
    - Go for 0-100 psi transmitters for hydro testing. Anything higher will increase the error of the measurements.
- Bucket and a stop watch
- Load Cell capable of giving mass vs time data would be even better

## Testing Procedures

- The full testing procedure should be its own document, this should be a rough summary. 

## Analysis of the results

### Loss Coefficients

- Convert volume flow rate of water in gpm or liters/min to mass flow rate in kg/s 
- convert pressure from psi to pa
- solve loss coefficient equation (7) using mass flow rate, pressure, density of water (measure the water temp, or estimate based on available data on typical tap water temp), and exit flow areas based on the measured as-built exit geometry.

### Spray Angle 

- measure the angle, adjust the flow until you get a good 45 degree spray, then do full tests to determine what the pressure drops for each flow path which create that angle.

### Momentum Ratio

- slightly more complicated to solve for, but with the pressure drops from the spray angle tests along with the loss coefficients from the loss coefficient tests, should be able to work out what the momentum ratio would be for actual propellants selected for engine design.
